```python
users = {}
for name, group in data2_with_tags.groupby("username"):
    users[name] = {}
    tags_set = set()
    for t in group["tags"]:
        tag_set = set(t.split("***"))
        tags_set.update(tag_set)
    users[name]["used_tags"] = len(tags_set)
    users[name]["posts"] = len(group["tags"])
```

```python
users["ansem_store"]
```

```python
users_df = pd.DataFrame.from_dict(users, orient="index")
```

```python
users_df.head()
```

```python
posts_used_tags = sns.lmplot('posts', 'used_tags', data=users_df, fit_reg=False, size=10)
```

```python
users_df[(users_df["used_tags"] < 200) & (users_df["posts"] > 100)]
```

```python
sns_plot = sns.kdeplot(users_df.posts, shade=True)
fig = sns_plot.get_figure()
fig.figsize=(18, 6)
```

```python
users_dates = {}
for name, group in data2_with_tags.groupby("username"):
    if len(group["local_date"]) > 1:
        users_dates[name] = (max(group["local_date"]) - min(group["local_date"])).days
    else:
        users_dates[name] = np.nan
```

```python
users_dates[list(users_dates.keys())[0]]
```

```python
users_dates_df = pd.DataFrame.from_dict(users_dates, orient="index")
```
