```python
users = {}
for name, group in data2_with_tags.groupby("username"):
    users[name] = {}
    tags_set = set()
    for t in group["tags"]:
        tag_set = set(t.split("***"))
        tags_set.update(tag_set)
    users[name]["used_tags"] = len(tags_set)
    users[name]["posts"] = len(group["tags"])
```

```python
users_df = pd.DataFrame.from_dict(users, orient="index")
```
